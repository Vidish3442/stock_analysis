```python
import fireducks.pandas as pd
import matplotlib.pyplot as plt
```

```python
# Step 1️: Load dataset
def load_data(filepath="indexData.csv"):
    print("Loading stock data into FireDucks...")
    data = pd.read_csv(filepath, parse_dates=["Date"])
    data.set_index("Date", inplace=True)
    return data
```

```python
# Step 2️: Compute moving averages
def compute_moving_averages(fd_data):
    print("Computing moving averages...")
    # Calculate 50-day and 200-day simple moving averages
    fd_data["50_MA"] = fd_data["Close"].rolling(window=50).mean()
    fd_data["200_MA"] = fd_data["Close"].rolling(window=200).mean()
    return fd_data
```

```python
# Step 3️: Visualize stock trends
def plot_stock_data(fd_data, name="Stock"):
    print("Plotting stock data...")
    plt.figure(figsize=(12,6))
    plt.plot(fd_data.index, fd_data["Close"], label="Close Price", color="blue")
    plt.plot(fd_data.index, fd_data["50_MA"], label="50-Day MA", linestyle="dashed")
    plt.plot(fd_data.index, fd_data["200_MA"], label="200-Day MA", linestyle="dotted")
    plt.legend()
    plt.title(f"{name} Close Price with Moving Averages")
    plt.xlabel("Date")
    plt.ylabel("Price")
    plt.grid(True)
    plt.show()
```

```python
# Step 4️: Run analysis
fd_data = load_data()
fd_data = compute_moving_averages(fd_data)
plot_stock_data(fd_data, name="Index Data")
print("Analysis Complete!")
```
